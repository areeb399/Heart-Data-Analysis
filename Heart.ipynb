```python
#Array containing all the unique values present in the 'Age' column 

Unique_Age = df['Age'].unique()

print("Unique Age : ", Unique_Age)
```

```python
#Number of unique values in the 'Age' column:

Count_Age = df['Age'].nunique()

print("Count of Unique Age : ", Count_Age)
```

```python
#Statistical summary of the 'Age' column:

df['Age'].describe()
```

```python
#Bins for various age categories:
young = df['Age'].quantile(0.25) #find the 25th percentile (Q1) of the 'Age' column 

mid_age = df['Age'].median() #find the median value in the 'Age' column

senior = df['Age'].quantile(0.75) #find the 75th percentile (Q3) value in the 'Age' column

elder = df['Age'].max() #find the maximum (highest) value in the 'Age' column
```

```python
# Define the labels for each bin:
bin_labels = ["Young_Adult", "Middle_Aged" , "Senior_Citizen", "Elderly"]

bins = [0, young, mid_age, senior, elder]
```

```python
# Create a new column 'Age_Category' based on the bins and labels

df['Age_Category'] = pd.cut(df['Age'], labels = bin_labels, bins = bins)
```

```python
# Display the DataFrame with the new 'Age_Category' column

print(df)
```

```python
#Creating a visualization on Age_Category Column 

fig = sns.countplot(x = "Age_Category" , data = df)


# Add count labels on top of each bar
for bars in fig.containers:
    fig.bar_label(bars)

# Set plot labels and title
plt.xlabel('Age Category', fontweight = 'bold')
plt.ylabel('Count', fontweight = 'bold')
plt.title('Count of People in Different Age Categories', fontweight = 'bold')

# Show the plot
plt.show()
```

```python
#Adding a new column 'Sex_Category' based on the values in the 'Sex' column

df['Sex_Category'] = np.where(df['Sex']==1, 'Male', 'Female')

#It assigns 'Male' to the 'Sex_Category' column if the corresponding value in the 'Sex' column is equal to 1, and 'Female' if it's 0
```

```python
#Adding a new column 'ExAng_Category' based on the values in the 'Exercise Induced Angina' column

df['ExAng_Category'] = np.where(df['Exercise Induced Angina'] == 1, 'Yes', 'No')

#It assigns 'Yes' to the 'ExAng_Category' column if the corresponding value in the 'Exercise Induced Angina' column is equal to 1, and 'No' if 0.
```

```python
#Creating a visualization on ChestPain Column  at different Age_Categories

# Set the figure size using plt.figure
plt.figure(figsize=(10, 6))

fig = sns.countplot(x = 'ChestPain', hue = 'Age_Category' , data = df)


# Add count labels on top of each bar
for bars in fig.containers:
    fig.bar_label(bars)

# Set plot labels and title
plt.xlabel('Chest Pain', fontweight = 'bold')
plt.xticks(rotation = 45)
plt.ylabel('Count', fontweight = 'bold')
plt.title('Chest Pain in Different Age Categories', fontweight = 'bold')

# Show the plot
plt.show()
```

```python
#Typical - Characteristic chest pain associated with heart-related issues like a heart attack or angina.

#Asymptomatic - Absence of chest pain, even in the presence of heart-related conditions.

#Nonanginal - Chest pain that is not related to the heart and does not fit the typical symptoms.

#Non-Typical - Chest pain with atypical characteristics, not fitting the classic heart-related symptoms.
```

```python
#Resting BP - BP when person at rest and not engaged in any physical activity or stress-inducing situations.
```

```python
#Creating a visualization on ChestPain Column with respect to Resting Blood Pressure

# Set the figure size using plt.figure
plt.figure(figsize=(10, 6))

fig = sns.boxplot(x = 'ChestPain', y = 'Resting Blood Pressure (mm Hg)' , data = df)

# Calculate the count for each category
count_data = df.groupby(['Resting Blood Pressure (mm Hg)', 'ChestPain']).size().reset_index(name='Count')



# Set plot labels and title
plt.xlabel('Chest Pain', fontweight = 'bold')
plt.xticks(rotation = 45)
plt.ylabel('Resting Blood Pressure(mm Hg)', fontweight = 'bold')
plt.title('Chest Pain with respect to Rest BP', fontweight = 'bold')

# Show the plot
plt.show()
```

```python
#Visualizations of above data

columns = ['ChestPain', 'Age_Category', 'Sex_Category', 'ExAng_Category']

def plot_cat_count():
    for c in columns :
        plt.figure(figsize=(10, 6))
        fig = sns.countplot(x = c, data = df, palette='dark')
        plt.title(c, fontweight = 'bold')
        plt.ylabel('Count', fontweight = 'bold')
        plt.xlabel(c, fontweight = 'bold')
        plt.tight_layout()
        
        # Add count labels on top of each bar
        for bars in fig.containers:
                fig.bar_label(bars)
       
        plt.show()
        
plot_cat_count()
```

```python
#Creating visualization of Maximum Heart Rate wrt Age Category for different Sex Category
plt.figure(figsize = (10,10))
fig = sns.catplot(x='Age_Category', y='Maximum Heart Rate', data=df, col='Sex_Category', kind='bar', height=5, aspect=1.2, palette='muted')

plt.suptitle('Maximum Heart Rate Age and Sex Categories', fontweight='bold')

plt.tight_layout()

plt.show()
```

```python
#Creating visualization of Maximum Heart Rate wrt Chest pain for different Sex Category

plt.figure(figsize = (10,10))
sns.catplot(x='ChestPain', y='Maximum Heart Rate', data=df, kind='bar', col='Sex_Category', height=5, aspect=1.2, palette='pastel')
plt.suptitle('Maximum Heart Rate by Chest Pain Categories and Sex Categories', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Creating visualization of Maximum Heart Rate wrt ExAng for different Sex Category

sns.catplot(x='ExAng_Category', y='Maximum Heart Rate', data=df, kind='violin', col='Sex_Category', height=5, aspect=1.2, palette='dark')
plt.suptitle('Maximum Heart Rate by ExAng Categories and Sex Categories', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Creating visualization of Rest BP wrt Sex Category for different Age Category

fig = sns.catplot(x='Sex_Category', y='Resting Blood Pressure (mm Hg)', data=df, col='Age_Category', kind='bar', height=5, aspect=1.2, palette='muted', col_wrap = 2)

plt.suptitle('Resting Blood Pressure in Different Age Categories', fontweight='bold')

plt.tight_layout()

plt.show()
```

```python
#Creating visualization of Cholestrol  wrt Sex Category for different Age Category

fig = sns.catplot(x='Sex_Category', y='Cholestrol (mg/dl)', data=df, col='Age_Category', kind='bar', height=5, aspect=1.2, palette='dark', col_wrap = 2)

plt.suptitle('Cholestrol in Different Age Categories', fontweight='bold')

plt.tight_layout()

plt.show()
```

```python
#Creating Visualization for Relationship Between Cholestrol and RestBP
sns.relplot(x='Cholestrol (mg/dl)', y='Resting Blood Pressure (mm Hg)', data=df, kind='scatter', hue='Sex_Category', col = 'Age_Category', height=6, aspect=1.2, palette='hsv',col_wrap=2)
plt.suptitle('Relationship Between Cholestrol and RestBP', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
Conclusion from above visualizations:

1. Predominantly, both male and female individuals reported experiencing typical chest pain,
   while a smaller proportion described non-typical symptoms.

2. On average, females exhibit a slightly higher heart rate than males. 

3. The blood pressure of females shows a higher increase with each passing year compared to males.

4. Over the years, female cholesterol levels show a higher increase compared to males.
```
